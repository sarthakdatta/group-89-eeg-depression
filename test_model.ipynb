```python
"""
1. Upload 'eeg_depression_model.keras'
2. Run this code
3. Output: Prediction (not depressed or depressed) and probability scores
"""
import tensorflow as tf
import numpy as np

model = tf.keras.models.load_model("eeg_depression_model.keras")
dummy_features = np.random.randn(1, 640)
prob = model.predict(dummy_features)[0, 0]
pred_0_or_1 = int(prob >= 0.5)
pred_label = "depressed" if pred_0_or_1 == 1 else "not depressed"

print(f"Probability of depression: {prob * 100:.1f}%")
print(f"Prediction: {pred_label}")
```
